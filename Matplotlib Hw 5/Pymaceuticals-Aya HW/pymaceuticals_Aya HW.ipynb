```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouseDrugData = "data/mouse_drug_data.csv"
clinicalTrialData = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouseData = pd.read_csv(mouseDrugData)
trialData = pd.read_csv(clinicalTrialData)


# Combine the data into a single dataset
mergedDf = trialData.merge(mouseData, on = 'Mouse ID')

# Display the data table for preview
mergedDf.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
avgVolume = pd.DataFrame(mergedDf.groupby(['Drug', 'Timepoint']).mean()['Tumor Volume (mm3)'])


# Preview DataFrame
avgVolume
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
volumeSE = pd.DataFrame(mergedDf.groupby(['Drug', 'Timepoint']).sem()['Tumor Volume (mm3)'])

# Preview DataFrame
volumeSE
```

```python
volumeSE = volumeSE.unstack(level=0)
volumeSE.columns = volumeSE.columns.get_level_values(1)

volumeSE
```

```python
# Minor Data Munging to Re-Format the Data Frames
avgVolume = avgVolume.unstack(level=0)
avgVolume.columns = avgVolume.columns.get_level_values(1)

# Preview that Reformatting worked
avgVolume
```

```python
# Generate the Plot (with Error Bars)
drugs =["Capomulin","Infubinol","Ketapril","Placebo"]
colors = ["red","blue", "green", "black"]
markers = ["o","^","s","d"]
fig, ax = plt.subplots()

for drug in drugs:
    xAxis = avgVolume.index.values
    yAxis = avgVolume[drug]
    err = (volumeSE[drug])
    
    index=drugs.index(drug)
    
    ax.errorbar(xAxis, yAxis, err, color=colors[index], marker=markers[index])

    ax.legend(loc="best", fancybox=True, numpoints=2, edgecolor="black")

plt.grid(linestyle="dotted")
ax.set_xlim(0,45)
ax.set_ylim(30,80)

ax.set_title("Tumor Response to Treatment", fontsize=14)
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Volume (mm3)")

# Save the Figure
plt.savefig("tumor_response_to_treatment")
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
summaryData = round(((avgVolume.iloc[9,:]-avgVolume.iloc[0,:])*100/avgVolume.iloc[0,:]),6).to_frame()

summaryData=summaryData.rename(columns={0:"% Tumor Change"})

# Display the data to confirm
summaryData
```

```python
# Store all Relevant Percent Changes into a Tuple
xAxis = summaryData.index.values
yAxis = summaryData["% Tumor Change"]

# Splice the data between passing and failing drugs
summaryData["Positive"]=summaryData["% Tumor Change"]>0

# Orient widths. Add labels, tick marks, etc. 
fig,ax=plt.subplots()

bars=ax.bar(xAxis, yAxis, align="edge", width=1, linewidth=1, 
            color=summaryData.Positive.map({True: 'green', 
                                                False: 'red'}))
ticks = [value+0.49 for value in range(len(xAxis))]

# Use functions to label the percentages of changes
height = summaryData.Positive.map({True: 5, False: -5})

# Call functions to implement the function calls
for bar in bars:
    ax.text(bar.get_x() + bar.get_width()/2., height[bars.index(bar)],
    str(int(summaryData["% Tumor Change"][bars.index(bar)])) + "%", 
    ha='center', va='bottom')

ax.set_xlim(0, len(xAxis))
plt.hlines(0,0,len(xAxis))
plt.grid(linestyle='dotted')

ax.set_title("Tumor Change over 45 Day Treatment", fontsize=14)
ax.set_ylabel("% Tumor Volume Change")
ax.set_xticks(ticks)
plt.setp(ax.get_xticklabels(), rotation=45)

# Save the Figure
#plt.savefig("Tumor_Change_Over_45_Day_Treatment")

# Show the Figure
plt.show()
```
